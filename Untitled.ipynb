```python
# Rozk≈Çady do dopasowania
distributions = ['gumbel_r', 'genextreme', 'pearson3']
```

```python
result = {
    'sezon': [],
    'rozk≈Çad': [],
    'loc': [],
    'scale': [],
    'shape': [],
    'aic': [],
    'woda stulecia': [],
    'woda piƒôcsetlecia': []
}

for dist_name in distributions:
    dist = getattr(sp.stats, dist_name)
    params_lato = dist.fit(przeplyw_max_lato['przeplyw'])
    params_zima = dist.fit(przeplyw_max_zima['przeplyw'])
    params_rok = dist.fit(przeplyw_max['przeplyw'])
    
    # Parametry dla lata
    args_lato = params_lato[:-2]
    loc_lato = params_lato[-2]
    scale_lato = params_lato[-1]
    aic_lato = 2 * len(args_lato) + 2 * dist.nnlf(params_lato, przeplyw_max_lato['przeplyw'])
    sto_lato = dist.ppf(0.99, loc = loc_lato, scale = scale_lato, *args_lato)
    pincset_lato = dist.ppf(0.995, loc = loc_lato, scale = scale_lato, *args_lato)
    
    # Parametry dla zimy
    args_zima = params_zima[:-2]
    loc_zima = params_zima[-2]
    scale_zima = params_zima[-1]
    aic_zima = 2 * len(args_zima) + 2 * dist.nnlf(params_zima, przeplyw_max_zima['przeplyw'])
    sto_zima = dist.ppf(0.99, loc = loc_zima, scale = scale_zima, *args_zima)
    pincset_zima = dist.ppf(0.995, loc = loc_zima, scale = scale_zima, *args_zima)
    
    # Parametry dla ca≈Çego roku
    args_rok = params_rok[:-2]
    loc_rok = params_rok[-2]
    scale_rok = params_rok[-1]
    aic_rok = 2 * len(args_rok) + 2 * dist.nnlf(params_rok, przeplyw_max['przeplyw'])
    sto_rok = dist.ppf(0.99, loc = loc_rok, scale = scale_rok, *args_rok)
    pincset_rok = dist.ppf(0.995, loc = loc_rok, scale = scale_rok, *args_rok)
    
    # Zapisz wyniki
    # Lato
    result['sezon'].append('lato')
    result['rozk≈Çad'].append(dist_name)
    result['loc'].append(loc_lato)
    result['scale'].append(scale_lato)
    if len(params_lato) == 3:
        result['shape'].append(params_lato[0])
    else:
        result['shape'].append(None)
        
    result['aic'].append(aic_lato)
    result['woda stulecia'].append(sto_lato)
    result['woda piƒôcsetlecia'].append(pincset_lato)
    
    # Zima
    result['sezon'].append('zima')
    result['rozk≈Çad'].append(dist_name)
    result['loc'].append(loc_zima)
    result['scale'].append(scale_zima)
    if len(params_lato) == 3:
        result['shape'].append(params_zima[0])
    else:
        result['shape'].append(None)
        
    result['aic'].append(aic_zima)
    result['woda stulecia'].append(sto_zima)
    result['woda piƒôcsetlecia'].append(pincset_zima)
    
    # Rok
    result['sezon'].append('rok')
    result['rozk≈Çad'].append(dist_name)
    result['loc'].append(loc_rok)
    result['scale'].append(scale_rok)
    
    if len(params_lato) == 3:
        result['shape'].append(params_rok[0])
    else:
        result['shape'].append(None)
    result['aic'].append(aic_rok)
    result['woda stulecia'].append(sto_rok)
    result['woda piƒôcsetlecia'].append(pincset_rok)
    
result = pd.DataFrame(result)
```

Reparametryzacja rozk≈Çadu:
$$
    \mu = \alpha * t + \beta
$$
oraz
$$
    \sigma = \gamma * t + \delta.
$$

Gƒôsto≈õƒá rozk≈Çadu GEV:
$$
    f(x) = \frac{1}{\sigma} t(x)^{\xi + 1} e^{-t(x)},
$$
gdzie
$$
    t(x) = \begin{cases}
        \left(1 + \xi \cdot \left(\frac{x - \mu}{\sigma}\right)\right)^{-\frac{1}{\xi}} & \xi \neq 0 \\
        e^{-\frac{x - \mu}{\sigma}} & \xi = 0
    \end{cases}
$$

### Teoretyczne warto≈õci liniowych moment√≥w
$$
    \lambda_1 = \mu - \sigma \frac{1 - \Gamma(1 - \xi)}{\xi}
$$
oraz
$$
    \lambda_2 = \sigma (1 - 2 ^ \xi) \frac{1 - \Gamma(1 - \xi)}{\xi}
$$
oraz
$$
    \tau_3 = 2 \frac{1 - 3 ^ \xi}{1 - 2^ \xi} - 3
$$
dla $x > \mu + \sigma / \xi$.

```python
# Wyznaczanie trendu liniowego
trend = przeplyw_max['przeplyw'].expanding().mean().to_numpy().reshape((-1, 1))
odchylenie = przeplyw_max['przeplyw'].expanding().std().to_numpy()[1:].reshape((-1, 1))
x = np.arange(przeplyw_max['rok_hydro'].min(), przeplyw_max['rok_hydro'].max() + 1, 1).reshape((-1, 1))
```

```python
lm = LinearRegression()
res = lm.fit(x, trend)
mu_0 = float(res.intercept_)
mu_1 = float(res.coef_)

print(f"Trend dla ≈õredniej: mu(t) = {mu_1:.3f} * t {mu_0:.3f}")
```

```python
lm = LinearRegression()
res = lm.fit(x[1:, :], odchylenie)
sigma_0 = float(res.intercept_)
sigma_1 = float(res.coef_)

print(f"Trend dla odchylenia: sigma(t) = {sigma_1:.3f} * t {sigma_0:.3f}")
```

```python
sns.lineplot(y='przeplyw', x='rok_hydro', data=przeplyw_max)
plt.plot(przeplyw_max['rok_hydro'], przeplyw_max['przeplyw'].expanding().mean())
plt.plot(x, mu_1 * x + mu_0)
plt.title('Przep≈Çyw maksymalny wody - dane roczne')
```

```python
# Dane wystandaryzowane
przeplyw_max_stand = przeplyw_max.iloc[1:, :]
przeplyw_max_stand['przeplyw_stand'] = (przeplyw_max_stand['przeplyw'] - (mu_1 * x.ravel()[1:] + mu_0)) / (sigma_1 * x.ravel()[1:] + sigma_0)
```

```python
sns.lineplot(x='rok_hydro', y='przeplyw_stand', data=przeplyw_max_stand)
plt.title('Przep≈Çyw roczny - usuniƒôty trend')
```

```python
l_moments = distr.gev.lmom_fit(przeplyw_max_stand['przeplyw_stand'])
loc = l_moments['loc']
scale = l_moments['scale']
shape = l_moments['c']
print(loc, scale, shape)
```

```python
q_100 = sp.stats.genextreme(loc=loc, scale=scale, *[shape]).ppf(0.99)
q_500 = sp.stats.genextreme(loc=loc, scale=scale, *[shape]).ppf(0.995)
```

```python
sns.lineplot(y='przeplyw', x='rok_hydro', data=przeplyw_max)
plt.plot(x[1:], q_100 * (sigma_1 * x.ravel()[1:] + sigma_0) + (mu_1 * x.ravel()[1:] + mu_0))
plt.plot(x, np.repeat(result.iloc[5, 6], len(x)))
plt.title('Woda stulecia - przep≈Çyw roczny z trendem')
```

```python
sns.lineplot(y='przeplyw', x='rok_hydro', data=przeplyw_max)
plt.plot(x[1:], q_500 * (sigma_1 * x.ravel()[1:] + sigma_0) + (mu_1 * x.ravel()[1:] + mu_0))
plt.plot(x, np.repeat(result.iloc[5, 7], len(x)))
plt.title('Woda piƒôcsetlecia - przep≈Çyw roczny z trendem')
```
